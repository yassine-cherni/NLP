```python
def extract_mfcc(wav_file_name):
    y, sr = librosa.load(wav_file_name)
    mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)

    return mfccs
```

```python
label_dict = {
    "DISABLE": 0,
    "LANE": 1,
    "ASSIST": 2,
    "SPEED": 3,
    "LIMIT": 4,
    "SET": 5,
    "KPH": 6,
}

HOUSSEM_loading = []
HOUSSEM_speech_data = []
HOUSSEM_speech_labels = []

for dirname, _, filenames in os.walk('/content/drive/MyDrive/DATA'):
    for filename in filenames:
        # Extract the label from the filename using the dictionary
        label = label_dict[filename.split('.')[0].split(' ')[2]]
        HOUSSEM_speech_labels.append(label)
        wav_file_name = os.path.join(dirname, filename)
        HOUSSEM_speech_data.append(extract_mfcc(wav_file_name))

print("Finish Loading the Dataset")
```
